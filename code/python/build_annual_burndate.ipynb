```python
import arcpy
from arcpy.sa import *
import os, glob

arcpy.env.overwriteOutput = True

maindir = "C:/Users/Maxwell/OneDrive/mcook/earth-lab/fastest-fires/data/spatial/raw/mcd64a1/"
outdir = "C:/Users/Maxwell/OneDrive/mcook/earth-lab/fastest-fires/data/spatial/mod/mcd64a1/"
dirs = os.listdir(maindir)
print(dirs)

for d in dirs:
    path = os.path.join(maindir, d)
    print(path)
    out_file = "mcd64a1_annual_burndate_y"+str(d)+".tif"
    if not os.path.exists(os.path.join(path, out_file)):
        rasters = glob.glob(path+"/*burndate.tif", recursive=True)
        print([os.path.basename(rast) for rast in rasters])
        out_rasters = []
        for r in rasters:
            OutRas = SetNull(r, r, "Value <= 0")
            out_rasters.append(OutRas)
        print("Number of Extracted Rasters: "+str(len(out_rasters)))
        print("Mosaicing for burn year: "+str(d))
        mosaic = arcpy.management.MosaicToNewRaster(
            out_rasters, outdir, out_file, 
            "", "16_BIT_UNSIGNED", "", 1, "Last", "")
    else:
        pass
```
